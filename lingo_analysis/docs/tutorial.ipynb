# 开始

## 打印原始文档

可以通过`raw`函数输出文档信息

````python
o.raw()
````

可以通过参数控制行数

````python
o.raw(50) # 打印文件前50行
````

## 查看基本信息

````python
print(o)
````

以上信息存储在字典`info`中

````python
o.info
````

## 查看变量信息

变量存储在字典`variable`中,可以直接查看

````python
o.variable
````

因为是字典, 可以转化为`DataFame`查看

````python
import pandas as pd
df = pd.DataFrame(o.variable)
df.head()
````

````python
d = pd.DataFrame(o.variable["D"])
d
````

### 访问单个变量信息

单个变量可以直接通过以下方式访问,这样访问不区分大小写

````python
o['x']
````

它等效于

````python
o.variable['X']
````

这种访问方式也可以获得基本信息

````python
o[0]
````

### 访问决策变量

在整数规划中,会有0-1的变量,当他们值为1是被储存到字典`decision`中

````python
o.decision
````

也可以通过重载的运算符`[1]`进行访问

````python
o[1]
````
